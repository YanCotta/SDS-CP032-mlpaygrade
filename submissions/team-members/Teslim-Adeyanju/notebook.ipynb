### 2.1 [Dataset Overview](#21-dataset-overview)  
____


The dataset used in this project is sourced from Kaggle and includes machine learning-related salary records for the year 2024. Each row represents an individual job role with attributes such as job title, location, experience level, employment type, company size, and more.

- **Source**: [Kaggle ‚Äì ML Engineer Salary (2024)](https://www.kaggle.com/datasets/chopper53/machine-learning-engineer-salary-in-2024)  
- **Observations**: ~16000+ records  
- **Format**: CSV  
- **Target Variable**: `salary_in_usd`  

This dataset provides a strong foundation for understanding salary distributions and building predictive models.

### 2.2 [Data Fields and Definitions](#22-data-fields-and-definitions)  



| Column Name         | Description                                                                 |
|---------------------|-----------------------------------------------------------------------------|
| `job_title`         | Title of the job (e.g., Data Scientist, ML Engineer)                        |
| `salary_in_usd`     | Annual salary in USD (standardized across all countries)                    |
| `employee_residence`| Country of the employee's primary residence                                 |
| `experience_level`  | Seniority level (EN = Entry, MI = Mid, SE = Senior, EX = Executive)         |
| `employment_type`   | Full-time, Part-time, Freelance, or Contract                                |
| `company_size`      | Size of the company (S = Small, M = Medium, L = Large)                      |
| `remote_ratio`      | % of remote work (0 = Onsite, 50 = Hybrid, 100 = Fully Remote)              |
| `company_location`  | Country where the company is headquartered                                  |
| `work_year`         | Year the salary was paid (should be 2024 for all entries)                   |



### 2.3 [Assumptions and Limitations](#24-assumptions-and-limitations)

**Assumptions**
- `salary_in_usd` is assumed to be standardized across all countries regardless of local currency.
- The dataset is considered representative of the global ML job market for the year 2024.
- High-cardinality categorical features such as `job_title` can be effectively captured using embedding layers in a deep learning model.

**Limitations**
- Potential **regional bias** due to overrepresentation from countries like the USA and India.
- The dataset lacks **industry/sector information** (e.g., tech, finance, healthcare), which may impact salary variance.
- **Total compensation** may be underestimated due to missing data on bonuses, equity, or benefits.
- **Temporal limitation**: The dataset only covers the year 2024 and does not account for salary trends over time.

```python
df['company_size'] = df['company_size'].cat.rename_categories({
    'S': 'Small',
    'M': 'Medium',
    'L': 'Large'
})  
```

```python
# Calculate key statistics
salary_stats = df['salary_in_usd'].describe()
mean_salary = salary_stats['mean']
median_salary = salary_stats['50%']
std_dev = salary_stats['std']
q1, q3 = salary_stats['25%'], salary_stats['75%']
iqr = q3 - q1

# Calculate outliers
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
outliers = df[(df['salary_in_usd'] < lower_bound) | (df['salary_in_usd'] > upper_bound)]
outlier_pct = (len(outliers) / len(df)) * 100

# Create comprehensive visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('Salary Distribution in USD - Comprehensive Analysis', fontsize=16, fontweight='bold')

# 1. Main Distribution with Key Metrics
sns.kdeplot(df['salary_in_usd'], ax=ax1, color='#3498DB', fill=True, alpha=0.4, linewidth=2)
ax1.set_title('Salary Distribution with Key Statistics', fontweight='bold', fontsize=12)
ax1.set_xlabel('Salary in USD')
ax1.set_ylabel('Density')

# Add vertical lines and annotations
metrics = [
    ('Mean', mean_salary, '#E74C3C'),
    ('Median', median_salary, '#2ECC71'),
    ('Q1 (25%)', q1, '#F39C12'),
    ('Q3 (75%)', q3, '#F39C12'),
    ('Outlier Threshold', upper_bound, '#9B59B6')
]

for label, value, color in metrics:
    ax1.axvline(value, color=color, linestyle='--', linewidth=1.5)
    ax1.text(value, ax1.get_ylim()[1]*0.1, f'{label}\n${value:,.0f}', 
             ha='center', va='bottom', fontweight='bold', color=color)

# Highlight IQR
ax1.axvspan(q1, q3, color='orange', alpha=0.1, label='IQR (Middle 50%)')

# 2. Boxplot with Outliers
sns.boxplot(x=df['salary_in_usd'], ax=ax2, color='#3498DB', width=0.4)
ax2.set_title('Boxplot Showing Outliers', fontweight='bold', fontsize=12)
ax2.set_xlabel('Salary in USD')

# Add annotations
ax2.text(0.5, 0.9, f'Outliers: {len(outliers):,} ({outlier_pct:.1f}% of data)', 
         transform=ax2.transAxes, ha='center', fontweight='bold',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='red'))

# 3. Cumulative Distribution
sns.kdeplot(df['salary_in_usd'], ax=ax3, color='#3498DB', cumulative=True, linewidth=2)
ax3.set_title('Cumulative Salary Distribution', fontweight='bold', fontsize=12)
ax3.set_xlabel('Salary in USD')
ax3.set_ylabel('Cumulative Probability')

# Add percentile markers
percentiles = [10, 25, 50, 75, 90]
for p in percentiles:
    value = df['salary_in_usd'].quantile(p/100)
    ax3.axvline(value, color='#E74C3C', linestyle=':', linewidth=1)
    ax3.text(value, p/100, f'{p}%: ${value:,.0f}', 
             ha='left', va='center', fontweight='bold', color='#E74C3C')
    ax3.axhline(p/100, color='gray', linestyle=':', linewidth=0.5)

# 4. Salary Brackets Analysis
bins = [0, 50000, 100000, 150000, 200000, 250000, 300000, 400000, 500000, float('inf')]
labels = ['<50k', '50-100k', '100-150k', '150-200k', '200-250k', '250-300k', '300-400k', '400-500k', '500k+']
df['salary_bracket'] = pd.cut(df['salary_in_usd'], bins=bins, labels=labels)
bracket_counts = df['salary_bracket'].value_counts().sort_index()

bracket_counts.plot(kind='bar', ax=ax4, color='#3498DB', width=0.7)
ax4.set_title('Salary Bracket Distribution', fontweight='bold', fontsize=12)
ax4.set_xlabel('Salary Bracket (USD)')
ax4.set_ylabel('Number of Employees')
ax4.tick_params(axis='x', rotation=45)

# Add percentage labels
total = len(df)
for i, count in enumerate(bracket_counts):
    pct = (count / total) * 100
    ax4.text(i, count + 100, f'{pct:.1f}%', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# Print key statistics
print("\nSalary Distribution Key Insights")
print("=" * 55)
print(f"1. Central Tendency:")
print(f"   - Mean salary: ${mean_salary:,.2f}")
print(f"   - Median salary: ${median_salary:,.2f} (Difference: ${mean_salary-median_salary:,.2f})")
print(f"\n2. Spread:")
print(f"   - Standard deviation: ${std_dev:,.2f}")
print(f"   - IQR (Middle 50% range): ${q1:,.2f} to ${q3:,.2f}")
print(f"\n3. Outliers:")
print(f"   - Upper threshold: ${upper_bound:,.2f}")
print(f"   - {len(outliers):,} outliers detected ({outlier_pct:.1f}% of data)")
print(f"\n4. Distribution Shape:")
print(f"   - Right-skewed (mean > median)")
print(f"   - {df['salary_in_usd'].skew():.2f} skewness, {df['salary_in_usd'].kurtosis():.2f} kurtosis")
print(f"\n5. Key Percentiles:")
for p in [10, 25, 50, 75, 90, 95, 99]:
    print(f"   - {p}% percentile: ${df['salary_in_usd'].quantile(p/100):,.2f}")
```
